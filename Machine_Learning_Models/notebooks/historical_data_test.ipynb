```python
import pandas as pd
import yfinance as yf
import plotly.express as px
```

```python
ticker = 'SPY'

ticker_data = yf.Ticker(ticker)
```

```python
start_date = '2018-01-01'

historical_data = ticker_data.history(start=start_date)

px.line(historical_data, x=historical_data.index, y='Close', title=f'{ticker} Price History Since {start_date}').show()

## Print(ticker_data)
```
